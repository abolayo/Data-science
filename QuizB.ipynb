```python
#import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv')
```

```python
df.head()
```

```python
columns = {
    'T1': 'Temperature_kitchen','RH_1':'Hum_kitchen','T2': 'Temp_L_room','RH_2':'Hum_L_room','T3': 'Temp_laundry','RH_3':'Hum_laundry',

'T4':'Temp_office','RH_4':'Hum_office','T5':'Temp_b_room','RH_5':'Hum_b_room','T6':'Temp_outside_NS','RH_6':'Hum_outside_NS',

'T7':'Temp_ironing_room','RH_7':'Hum_ironing_room','T8':'Temp_teenager_room','RH_8': 'Hum_teenager_room','T9':'Temp_parents_room',

'RH_9': 'Hum_parents_room','T_out':'Temperature_outside','rv1': 'Random_variable1','rv2':'Random_variable2'
}
df = df.rename(columns=columns)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop(['date', 'lights'],axis=1)
```

```python
df.head()
```

```python
# Doing MinMax scaling for Normalization
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalize_df = pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
features_df = normalize_df.drop(columns=['Appliances'])
appliances_df = normalize_df['Appliances']
appliances_df.head()
```

```python
# Using a 70-30 train-test set split with a random state of 42 (for reproducibility).
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df,appliances_df, test_size=0.3, random_state=42)
```

```python
from sklearn import metrics
from sklearn.linear_model import LinearRegression,Ridge,Lasso
linear_model = LinearRegression()
linear_model.fit(x_train,y_train)
predicted_value_1=linear_model.predict(x_test)


```

```python
predicted_value_1
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mae_1 = mean_absolute_error(y_test,predicted_value_1)
print('MEAN ABSOLUTE ERROR USING LINEAR REG:', mae_1)

mse_1 = metrics.mean_squared_error(y_test,predicted_value_1)
print('Mean Square Error for LINEAR REG',mse_1) 

rss_1 = np.sum(np.square(y_test - predicted_value_1))
print('RSS for LINEAR REG:',rss_1)

rmse_1 = np.sqrt(mean_squared_error(y_test,predicted_value_1))
print('RMSE for LINEAR REG:',rmse_1)

r2_score_1 =  metrics.r2_score(y_test,predicted_value_1)
print('RSQUARE for LINEAR REG :',r2_score_1)
```

```python
ridge_reg = Ridge(alpha=0.5)
ridge_reg.fit(x_train,y_train)
predicted_value_2 = ridge_reg.predict(x_test)
```

```python
mae_2 = mean_absolute_error(y_test,predicted_value_2)
print('MEAN ABSOLUTE ERROR USING RIDGE:', mae_2)

mse_2 = metrics.mean_squared_error(y_test,predicted_value_2)
print('Mean Square Error for RIDGE',mse_2) 

rss_2 = np.sum(np.square(y_test - predicted_value_2))
print('RSS for RIDGE:',rss_2)

rmse_2 = np.sqrt(mean_squared_error(y_test,predicted_value_2))
print('RMSE for RIDGE:',rmse_2)

r2_score_2 =  metrics.r2_score(y_test,predicted_value_2)
print('RSQUARE for  RIDGE :',r2_score_2)
```

```python
ridge_reg = Ridge(alpha=5)
ridge_reg.fit(x_train,y_train)
predicted_value_2 = ridge_reg.predict(x_test)
```

```python
mae_2 = mean_absolute_error(y_test,predicted_value_2)
print('MEAN ABSOLUTE ERROR USING RIDGE:', mae_2)

rss_2 = np.sum(np.square(y_test - predicted_value_2))
print('RSS for RIDGE:',rss_2)

mse_2 = metrics.mean_squared_error(y_test,predicted_value_2)
print('Mean Square Error for RIDGE',mse_2) 

rmse_2 = np.sqrt(mean_squared_error(y_test,predicted_value_2))
print('RMSE for RIDGE:',rmse_2)

r2_score_2 =  metrics.r2_score(y_test,predicted_value_2)
print('RSQUARE for  RIDGE :',r2_score_2)
```

```python
lasso_reg = Lasso(alpha=0.001) 
lasso_reg.fit(x_train, y_train)
predicted_value_3 = lasso_reg.predict(x_test)
```

```python
mae_3 = mean_absolute_error(y_test,predicted_value_3)
print('MEAN ABSOLUTE ERROR USING RIDGE:', mae_3)

rss_3 = np.sum(np.square(y_test - predicted_value_3))
print('RSS for LASSO:',rss_3)

mse_3 = mean_squared_error(y_test,predicted_value_3)
print('Mean Square Error for LASSO',mse_3) 

rmse_3 = np.sqrt(mean_squared_error(y_test,predicted_value_3))
print('RMSE for LASSO:',rmse_3)

r2_score_3 =  metrics.r2_score(y_test,predicted_value_3)
print('RSQUARE for  LASSO :',r2_score_3)
```

```python
#comparing the effects of regularisation 
def get_weights_df(model, feat, col_name):
#this function returns the weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index() 
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight') 
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight') 
```

```python
final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features') 
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
final_weights
```
